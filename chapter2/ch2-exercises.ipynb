```python
# a)
a = -6.3
v = 18/3.6
x = 34
# Don't know t, want to know v0
# Choose v^2 = v0^2 + 2*a*x
v0 = math.sqrt(v**2 - 2*a*x)
print(f'The motorist was going {v0:.1f} m/s when he started breaking.')
```

```python
# b) 
# Still don't know acceletation
# Simplest way: Use x = 0.5*(v0+v)*t
t = (2*x)/(v0+v)
print(f'Breaking took {t:.1f} seconds.')
```
